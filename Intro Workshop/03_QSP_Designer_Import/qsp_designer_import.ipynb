```python
import abm
import pandas as pd

parameters = pd.DataFrame(columns=["parameter", "value", "unit"], data=[["DoseSC", 50.0, "mg"], ["nDosesSC", 20.0, "1"]])
outputs = ["C_P_Drug"]
times = abm.linspace(0, 300, 501, "d")

simers = abm.simulate(models="MinPBPK_Tumour_IO.qsp", parameters=parameters, outputs=outputs, times=times)

print(simers.to_data_frame(tall_outputs=False))
```

# QSP Designer Import

## QSP Designer Model



Simple 2-compartment model in QSP Designer, a 1 mg bolus dose `D1` is administered to `V_1` at `t = 0` and simulated for 24 hours.




Model is saved as `basic_model.qsp`

## Running in Notebook

### Basic Model Run

We can run the model in Notebook using the normal `simulate` interface. The Designer model is passed using the `models` argument.

```python
# As in Designer, model is simulated for 24 hours at 200 timepoints
times = abm.linspace(0, 24, 200, "h")
sim = abm.simulate(models="basic_model.qsp", times=times)
```
