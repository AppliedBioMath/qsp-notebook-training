```python
from plotnine import *
import pandas as pd
import numpy as np
```

```python
# Set up some fake data
Tdat = pd.DataFrame({'t':[0,1,2,3], 'pk':[100, 10, 1, 0.1]})

# Set the LLoQ value
lloq = 0.3
```

```python
(
    ggplot(Tdat, aes('t', 'pk')) +
    # geom_line plots the PK line from the data
    geom_line() +
    # geom_hline adds a horizontal line
    geom_hline(
        yintercept = lloq, # yintercept sets the y value of the horizontal line
        linetype = "dashed" # linetype sets the dash type of the line
    ) +
    # geom_text adds a descriptive label
    geom_text(
        x = -np.inf, # x position of the text, -np.inf sets all the way to the left, np.inf sets all the way to the right
        y = np.log10(lloq), # y position of the text. Have to provide log of the value when using a log scale
        label = "LLoQ", # Text to show
        ha = "left", # horizontal alignment -- "left" sets the x value as the left side of the text
        va = "bottom", # vertical alignment = "bottom" sets the y value as the bottom of the text
        size = 14
    ) +
    scale_y_log10() +
    theme(
        figure_size=(4,3), # Sets the figure size to 4x3 in
        dpi = 100 # Sets the resolution (in dpi) to 100. Higher values are sharper
    )
)
```
